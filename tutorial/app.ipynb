# Use the similarity edge feature

We print all similar pairs of words that are at least 50% similar but not 100%.

```python
query = """
word
<sim>50> word
"""
```

```python
results = A.search(query)
```

```python
A.table(results, end=10, withPassage="1 2")
```

```python
A.show(results, end=5)
```
